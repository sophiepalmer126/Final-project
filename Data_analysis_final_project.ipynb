# Functions

```python
def get_r2(x,y):
    polyfit = np.polyfit(x,y,1)
    y_predicted = np.polyval(polyfit,x)
    residuals = y-y_predicted
    return 1-np.sum(residuals**2)/np.sum((y-np.mean(y))**2)
```

# Data

```python
data = pd.read_csv('/Users/sophiepalmer/Desktop/Data_Analysis_VIEPS/Final_Project/Finalised_C_G_BD_withcoordinates.csv')

C_av =data['%C av']
G_av =data['%G av']
```

# Analysis 

```python
#assesing outliers
plt.subplot(221)
plt.boxplot(C_av)

plt.subplot(222)
plt.boxplot(G_av)
# three outliers identified from this 
```

```python
Q1 = C_av.quantile(0.25)
Q3 = C_av.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (C_av < lower_bound) | (C_av > upper_bound)
#print(outliers)
```

```python
#filter out the three outlier values (identified above row 19,27,28)
data_filtered = data.drop([18,26,27])
C_filtered_av = data_filtered['%C av']
G_filtered_av = data_filtered['%G av']
```

## R2

```python
r2 =get_r2(C_filtered_av,G_filtered_av)
print(r2)
```

## Line of Best Fit

```python
coefficients = np.polyfit(C_filtered_av, G_filtered_av, 1)
line_bf = np.poly1d(coefficients)
x_values = np.linspace(min(C_filtered_av), max(C_filtered_av),100)
y_values = line_bf(x_values)
equation = f'y = {line_bf.coefficients[1]:.2f}x + {line_bf.coefficients[0]:.2f}'
print(equation)
```

# Plots 

```python
plt.scatter(C_filtered_av,G_filtered_av, zorder=3)
plt.plot(x_values, y_values, color='green', label='Line of Best Fit')
plt.xlabel('Organic Carbon (%)')
plt.ylabel('Gypsum (%)')

plt.text(0.05,0.9, equation, transform=plt.gca().transAxes, fontsize=10, bbox=dict(facecolor='white', 
            edgecolor='black', boxstyle='round,pad=0.5'))
plt.text(0.05,0.8, r"$R^2 = {:.3f}$".format(r2),  transform=plt.gca().transAxes, fontsize=10, bbox=dict(facecolor='white', 
            edgecolor='black', boxstyle='round,pad=0.5'))
plt.minorticks_on()
plt.grid(True,linestyle='--',c='lightgrey', zorder=0)
```

# Kriging 

## Upper 

```python
dataframe_data=pd.DataFrame(data)

#print(dataframe_data['%C_upper'])
mask = dataframe_data['%C_upper']<= 7
filtered_data = dataframe_data[mask]
```

```python
C_upper = filtered_data['%C_upper']
data_coordinates = filtered_data[['x', 'y']].values
C_upper_values  = C_upper.values
```

```python
okrig = OrdinaryKriging(
    data_coordinates[:, 0], 
    data_coordinates[:, 1], 
    C_upper_values, 
    variogram_model='linear', 
    enable_plotting=False)

grid_x, grid_y = np.mgrid[data['x'].min():data['x'].max():200j],np.mgrid[data['y'].min():data['y'].max():200j]
z_upper, ss_upper = okrig.execute('grid', grid_x, grid_y)
#plt.imshow(ss_upper)
```

```python
plt.contourf(grid_x, grid_y, z_upper, cmap='viridis',levels=45)
plt.colorbar()
plt.scatter(data_coordinates[:, 0], data_coordinates[:, 1], c='white', marker='o',s=2, label='Sample Sites')
plt.xlim([114.315,114.320])
plt.axis('tight')
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.ticklabel_format(useOffset=False)
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
```

## Lower
