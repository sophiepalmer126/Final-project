```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from osgeo import gdal
import xarray as xr
from scipy.io import loadmat
from pykrige import OrdinaryKriging
import shapefile as spf
from matplotlib.ticker import ScalarFormatter


%matplotlib inline
plt.rcParams['font.sans-serif'] = 'Helvetica' # Choose a nice font here
plt.rcParams['font.size'] = 7 # standard font size
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica'
plt.rcParams['figure.facecolor'] = 'white' # make sure background isn't transparent
plt.rcParams['figure.dpi'] = 200
```

## Read the Data

```python
data = pd.read_csv('/Users/sophiepalmer/Desktop/Data_Analysis_VIEPS/Final_Project/Finalised_C_G_BD_withcoordinates.csv')
dataframe_data=pd.DataFrame(data)

print(dataframe_data['%C_upper'])
mask = dataframe_data['%C_upper']<= 5
filtered_data = dataframe_data[mask]
```

## Upper

```python
C_upper = filtered_data['%C_upper']
data_coordinates = filtered_data[['x', 'y']].values
C_upper_values  = C_upper.values
```

```python
okrig = OrdinaryKriging(
    data_coordinates[:, 0], 
    data_coordinates[:, 1], 
    C_upper_values, 
    variogram_model='linear', 
    enable_plotting=True)

grid_x, grid_y = np.mgrid[data['x'].min():data['x'].max():200j],np.mgrid[data['y'].min():data['y'].max():200j]

z_upper, ss_upper = okrig.execute('grid', grid_x, grid_y)
#plt.imshow(ss_upper)
```

```python
plt.contourf(grid_x, grid_y, z_upper, cmap='viridis',levels=45)
#np.linspace
plt.colorbar()
plt.scatter(data_coordinates[:, 0], data_coordinates[:, 1], c='white', marker='o',s=2, label='Sample Sites')
plt.xlim([114.315,114.320])
plt.axis('tight')
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.ticklabel_format(useOffset=False)
```

## Lower 

```python
data_lower =  pd.read_csv('/Users/sophiepalmer/Desktop/Data_Analysis_VIEPS/Final_Project/C_lower_nospace.csv')

data_coordinates2 = data_lower[['x', 'y']].values

x_lower = data_lower['x']
y_lower = data_lower['y']
C_lower = data_lower['%C_lower']
```

```python
okrigL = OrdinaryKriging(
    data_coordinates2[:, 0], 
    data_coordinates2[:, 1], 
    C_lower, 
    variogram_model='linear', 
    enable_plotting=True)

grid_x2, grid_y2 = np.mgrid[data_lower['x'].min():data_lower['x'].max():200j],np.mgrid[data_lower['y'].min():data_lower['y'].max():200j]

z_lower, ss_lower = okrigL.execute('grid', grid_x2, grid_y2)
```

```python
plt.contourf(grid_x2, grid_y2, z_lower, cmap='viridis', levels=45)
plt.colorbar()
plt.xlim([114.315,114.320])
plt.axis('tight')
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.ticklabel_format(useOffset=False)
```
